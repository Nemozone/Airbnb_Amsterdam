## Import Modules:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load Data:

```python
df = pd.read_csv('listings.csv')
df.head()
```

## Picking the useful features:

```python
# droping all the unneccessary featyres such as names, urls, etc.
df_new = df[['id','neighbourhood_cleansed','latitude','longitude','property_type',
           'room_type','accommodates','amenities','price','number_of_reviews','review_scores_rating']]

# Reorder the columns in a way that price comes last as target for future modelings
column_reordered = ['id','neighbourhood_cleansed','latitude','longitude','property_type',
           'room_type','accommodates','amenities','number_of_reviews','review_scores_rating','price']
df_new = df_new[column_reordered]
df_new.head()
```

## Checking missing values and data types:

```python
# defining a fuction to point out columns with missing values:
def check_for_null(DF):
    null_cols = DF.columns[DF.isnull().any()]
    if null_cols.size == 0:
        print('There are no missing values')
    else:
        return null_cols   
            
```

```python
check_for_null(df_new)
```

```python
# As it turnes out, 'review_scores_rating' is the only column with missing values in out data frame
df_new_cleaned = df_new.dropna(subset=['review_scores_rating'], axis=0)
```

```python
check_for_null(df_new_cleaned)
```

```python
# checking the data types:
df_new_cleaned.dtypes
```

```python
# We see the 'price' type is object which needs to be numeric
df_new_cleaned['price'] = df_new_cleaned['price'].str.replace(',', '').str.replace('$', '').astype(float)
df_new_cleaned['price'].dtypes
```

## Question 1: In which neighbourhood it is more likely to find an affordable property?

```python
# first, we build a new df for more clarity:
df_q_1 = df_new_cleaned[['neighbourhood_cleansed', 'price']]

# in answer to our first question we should group our data based on neigbourhood:
df_q_1 = df_q_1.groupby(['neighbourhood_cleansed']).mean(['price'])
df_q_1['num_listings'] = df_new_cleaned['neighbourhood_cleansed'].value_counts()

# creating a new column called desirability to pointout neigbourhoods with more listings and less prices:
df_q_1['Desirability'] = df_q_1['num_listings'] / df_q_1['price']
df_q_1 = df_q_1.sort_values(by=['Desirability'], ascending=False)
df_q_1 = df_q_1.reset_index()

df_q_1
```

```python
df_q_1.plot(kind = 'bar', x= 'neighbourhood_cleansed', y= 'Desirability')
plt.title('Most Available and Afordable Areas', fontsize=20)
plt.xlabel('Neighbourhood', fontsize=16)
plt.ylabel('Desirability', fontsize=16)
plt.show()
```

Looking at the results, we see that the first two neighbours with the most desirability are: 'De Baarsjes' and 'De Pijp'. I am not familiar with Amsterdam city but taking a quick glance at the neighbourhood on google map, I can say that 'De Pijp' is cloder to the center and according to this analysis we have a high chance of finding affordable and availabe properties here.

## Question 2: Which Types of Properties are more expensive?

```python
df_q_2 = df_new_cleaned.groupby(['property_type']).mean(['price']).sort_values(by=['price'], ascending=False)
df_q_2 = df_q_2.reset_index()
df_q_2
```

## What is the level of satisfaction based on prices?

```python
# first we change the type of 'amenities' column from string to a list :
df_new_cleaned['amenities'] = df_new_cleaned['amenities'].apply(lambda x: x[1:-1].split(','))

# now we can easily count the elements inside it:
df_new_cleaned['num_of_amenities'] = df_new_cleaned['amenities'].str.len()
df_new_cleaned = df_new_cleaned[['id','neighbourhood_cleansed','latitude','longitude','property_type',
           'room_type','accommodates','amenities', 'num_of_amenities', 'number_of_reviews','review_scores_rating','price']]

# since we do not need the amenities column anymore, so we drop it and save it into a new dataframe:
df_cleaned = df_new_cleaned.drop(['amenities'], axis=1)
df_cleaned
```

```python
# now we should find an algoryhtm to calculate a fair score considering both ratings and number of ratings:
# I found this algorythm here: https://bit.ly/3lTGXpO
from math import e
p = df_cleaned['review_scores_rating']
q = df_cleaned['number_of_reviews']
scores = (0.5 * p) + 5 * (1-e**(-q / 10))
```

```python
# we add the score column we created to the dataframe and rearrange again
df_cleaned['Scores'] = scores
df_cleaned = df_cleaned[['id', 'neighbourhood_cleansed', 'latitude', 'longitude',
       'property_type', 'room_type', 'accommodates', 'num_of_amenities', 'Scores',
       'number_of_reviews', 'review_scores_rating', 'price']]

# drop the columns which we do not need anymore:
df_cleaned = df_cleaned.drop(['number_of_reviews', 'review_scores_rating'], axis=1)
```

```python
price_score = pd.DataFrame({'Score':df_cleaned['Scores'], 'Price':df_cleaned['price']})
price_score
```

```python

x = price_score['Score']
y = price_score['Price']

plt.scatter(x,y)
plt.xlabel('Score')
plt.ylabel('Price')

plt.show()
```

```python
# to deal with the outliers, we should get rid of the upper 1% of prices:
q = price_score['Price'].quantile(0.99)
price_score.drop(price_score[price_score['Price']>q].index, inplace=True)
```

```python
x = price_score['Score']
y = price_score['Price']

plt.scatter(x,y, alpha=0.1)
plt.xlabel('Score')
plt.ylabel('Price')

plt.show()
```

## Question 4: In which neighbourhood we are more likely to find a high score property at a fair cost?

```python
# we make our dataframe with all the neccessary data:
df_q_4 = df_cleaned.groupby(['neighbourhood_cleansed']).mean()
df_q_4 = df_q_4.reset_index()
df_q_4 = df_q_4.drop(['id','latitude', 'longitude', 'accommodates'], axis=1)
df_q_4
```

```python
# we normalize the price and score columns:
def normalizer(data):
    data_n=((data-data.min())/(data.max()-data.min()))*20
    return data_n

Price_scaled = normalizer(df_q_4['price'])
Score_scaled = normalizer(df_q_4['Scores'])

df_q_4_n = pd.DataFrame({'Neighbourhood':df_q_4['neighbourhood_cleansed'], 
                         'Score':Score_scaled, 'Price':Price_scaled}, )
df_q_4_n = df_q_4_n.sort_values(by=['Score'])
df_q_4_n
```

```python
x= df_q_4_n['Neighbourhood']
y1= df_q_4_n['Price']
y2= df_q_4_n['Score']
fig=plt.figure(figsize=(20,5)) 
ax = fig.add_subplot(111)
ax.plot(x, y1, '-b', label='Price')
ax.plot(x,y2, '-r', label='Score')
plt.xticks(x[::1], fontsize=16, rotation='vertical')
plt.title('Price - Score', fontsize=20)


leg = ax.legend()
plt.show()
```

According to the previous graph, we can easily spot that the highest score areas with affordable prices at 'Noord-West', 'Bos en Lommer', and 'De Aker-Nieuw Sloten'
